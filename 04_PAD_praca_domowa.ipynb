4. Rozdziel kolumnę tshirt category na 3 kolumny: tshirt_type, która będzie mówiła o tym czy jest to koszulka, bluza czy coś innego, tshirt_size, która będzie informowała o rozmiarze koszulki, i tshirt_colour, która będzie informowała o kolorze. Skorzystaj z metody apply.

```python
import pandas as pd
```

1.

```python
data_orders = pd.read_csv("orders.csv")
data_orders['order_date'] = pd.to_datetime(data_orders['order_date'], format="%Y-%m-%d")
data_orders['order_date'].sample(1)
```

2.

```python
data_orders['tshirt_category'].value_counts()
```

3.

```python
data_orders['tshirt_category'] = data_orders['tshirt_category'].apply(
    lambda item : item\
        .replace("Tshirt", "T-Shirt")\
        .replace("Wh T-Shirt M", "White T-Shirt M")\
        .replace("Bl T-Shirt M", "Black T-Shirt M")\
        .replace("Wh T-Shirt F", "White T-Shirt F")\
        .replace("Bl T-Shirt F", "Black T-Shirt F")\
        .replace("Tennis Shirt", "Unknown Tennis T-Shirt")\
        .replace("Hoodie", "Unknown Hoodie Unknown")
)

data_orders['tshirt_category'].value_counts()
```

4.

```python
data_orders['tshirt_colour'] = data_orders['tshirt_category'].apply(lambda item : item.split(" ", 3)[0])
data_orders['tshirt_type'] = data_orders['tshirt_category'].apply(lambda item : item.split(" ", 3)[1])
data_orders['tshirt_size'] = data_orders['tshirt_category'].apply(lambda item : item.split(" ", 3)[2])
data_orders
```

5.

```python
data_orders[(data_orders['order_date'] >= '2014-12-31') & (data_orders['order_date'] <= '2016-08-02')].sort_values(by='order_date')
```
